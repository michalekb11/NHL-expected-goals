### Insert player and goalie statistics into MySQL tables

```python
# Packages
import pandas as pd
from MySQLClass import MySQL
```

```python
# User input
# skater_path = '/Volumes/LUNANI/sports_betting_csv_backup/game_stats/skater_games.csv'
# goalies_path = '/Volumes/LUNANI/sports_betting_csv_backup/game_stats/goalie_games.csv'

skater_path = '/Users/bryanmichalek/Documents/GitHub_Personal/sports_betting_data/data/historic_batch/stats/skater_2.csv'
goalies_path = '/Users/bryanmichalek/Documents/GitHub_Personal/sports_betting_data/data/historic_batch/stats/goalie_2.csv'
```

```python
# Create the engine to connect to the MySQL database
mysql = MySQL()
mysql.connect()
```

### Skaters

```python
# Read in CSV files
skater = pd.read_csv(skater_path)
display(skater.head())
```

```python
# Select the correct columns, rename, and order them
skater = skater[['player_id', 'team', 'game_num', 'date', 'opponent', 'G', 'A', 'P', 'rating', 'PIM', 'EVG', 'PPG', 'SHG', 'GWG', 'EVA', 'PPA', 'SHA', 'S', 'S_perc', 'shifts', 'TOI', 'HIT', 'BLK', 'FOW', 'FOL', 'FOW_perc']].rename(columns={'S_perc':'S_pct', 'FOW_perc':'FOW_pct'})
```

```python
# Insert into skater statistics table
skater.to_sql(name = 'skater_game', con=mysql.engine, schema='nhl', if_exists='append', index=False)
```

### Goalies

```python
# Read in CSV files
goalie = pd.read_csv(goalies_path)
display(goalie.head())
```

```python
# Select the correct columns, rename, and order them
goalie = goalie[['player_id', 'team', 'game_num', 'date', 'opponent', 'decision', 'GA', 'SA', 'SV', 'SV_perc', 'shutout', 'PIM', 'TOI']].rename(columns={'SV_perc':'SV_pct'})
```

```python
# Insert into goalie statistics table
goalie.to_sql(name = 'goalie_game', con=mysql.engine, schema='nhl', if_exists='append', index=False)
```
