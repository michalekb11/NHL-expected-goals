### Figure out who in this training data was a C
Assign these players a 1 if they are a center. We will first need to investigate what constitutes being a center. The best indicator will be if a player consistently takes >= N number of faceoffs per game (or per 60) over the past few games (can use the rolling window from combined per 60). We need to figure out what the correct value for N should be. 

```python
per60_2021 = pd.read_csv('../data/stats/past_seasons/combined_per60_roll15_2021.csv')
per60_2022 = pd.read_csv('../data/stats/past_seasons/combined_per60_roll15_2022.csv')
```

```python
C_cutoff = 20.0
```

```python
feature_set_2021 = per60_2021.copy()
feature_set_2022 = per60_2022.copy()
```

```python
feature_set_2021['C_flag'] = 0
feature_set_2021.loc[feature_set_2021['FOL_60'] + feature_set_2021['FOW_60'] >= C_cutoff, 'C_flag'] = 1

feature_set_2022['C_flag'] = 0
feature_set_2022.loc[feature_set_2022['FOL_60'] + feature_set_2022['FOW_60'] >= C_cutoff, 'C_flag'] = 1
```

```python
pd.set_option('display.max_rows', None)
display(feature_set_2021.loc[feature_set_2021['C_flag'] == 1, 's_name'].value_counts().reset_index())
pd.set_option('display.max_rows', 10)
```

```python
pd.set_option('display.max_rows', None)
display(feature_set_2022.loc[feature_set_2022['C_flag'] == 1, 's_name'].value_counts().reset_index())
pd.set_option('display.max_rows', 10)
```

```python
print(feature_set_2021['C_flag'].value_counts())
print(feature_set_2022['C_flag'].value_counts())
```
