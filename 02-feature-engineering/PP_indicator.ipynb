```python
import pandas as pd
```

### Assign PP indicator feature to each row in training data using PP cutoff value from EDA

```python
feature_set_2021 = pd.read_csv('../data/feature_set_2021.csv')
feature_set_2022 = pd.read_csv('../data/feature_set_2022.csv')
```

```python
PP_cutoff = 0.35
```

```python
feature_set_2021['PP_flag'] = 0
feature_set_2021.loc[feature_set_2021['PPG_60'] + feature_set_2021['PPA_60'] >= PP_cutoff, 'PP_flag'] = 1

feature_set_2022['PP_flag'] = 0
feature_set_2022.loc[feature_set_2022['PPG_60'] + feature_set_2022['PPA_60'] >= PP_cutoff, 'PP_flag'] = 1
```
