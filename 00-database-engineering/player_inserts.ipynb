### Insert player and goalie statistics into MySQL tables

```python
# Packages
import pandas as pd
import sqlalchemy
```

```python
# User input
season = '2023'
skater_path = '/Volumes/LUNANI/sports_betting_csv_backup/stats/past_seasons/skaters_' + season + '.csv'
goalies_path = '/Volumes/LUNANI/sports_betting_csv_backup/stats/past_seasons/goalies_' + season + '.csv'
```

```python
# Create the engine to connect to the MySQL database
engine = sqlalchemy.create_engine('mysql+mysqlconnector://root:rootdata@localhost/nhl')
```

### Skaters

```python
# Read in CSV files
skater = pd.read_csv(skater_path)
display(skater.head())
```

```python
# Select the correct columns, rename, and order them
skater = skater[['player_id', 'player_name', 'age', 'team', 'game_num', 'date', 'opponent', 'G', 'A', 'P', 'rating', 'PIM', 'EVG', 'PPG', 'SHG', 'GWG', 'EVA', 'PPA', 'SHA', 'S', 'S_perc', 'shifts', 'TOI', 'HIT', 'BLK', 'FOW', 'FOL', 'FOW_perc']].rename(columns={'player_name':'name', 'S_perc':'S_pct', 'FOW_perc':'FOW_pct'})
```

```python
# Insert into skater statistics table
skater.to_sql(name = 'skater_games', con=engine, schema='nhl', if_exists='append', index=False)
```

### Goalies

```python
# Read in CSV files
goalie = pd.read_csv(goalies_path)
display(goalie.head())
```

```python
# Select the correct columns, rename, and order them
goalie = goalie[['player_id', 'player_name', 'age', 'team', 'game_num', 'date', 'opponent', 'decision', 'GA', 'SA', 'SV', 'SV_perc', 'shutout', 'PIM', 'TOI']].rename(columns={'player_name':'name', 'SV_perc':'SV_pct'})
```

```python
# Insert into goalie statistics table
goalie.to_sql(name = 'goalie_games', con=engine, schema='nhl', if_exists='append', index=False)
```
